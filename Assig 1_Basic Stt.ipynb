```python
# 1: Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import zscore
```

```python
#2: Load the dataset

df = pd.read_csv(r"C:\Users\USER\Downloads\sales_data_with_discounts.csv")

# Display the first few rows of the dataset
df.head()
```

```python
#  3: Identify numerical and categorical columns
numerical_cols = df.select_dtypes(include=[np.number]).columns.tolist()
categorical_cols = df.select_dtypes(include=['object', 'category']).columns.tolist()
```

```python
#  4: Descriptive Analytics for Numerical Columns
# Calculate mean, median, mode, and standard deviation for numerical columns
descriptive_stats = df[numerical_cols].describe().T
descriptive_stats['mode'] = df[numerical_cols].mode().iloc[0]

# Display descriptive statistics
descriptive_stats
```

```python
# 5: Data Visualization - Histograms for numerical columns
for col in numerical_cols:
    plt.figure(figsize=(8, 5))
    sns.histplot(df[col], kde=True, bins=30)
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

```python
# 6: Data Visualization - Boxplots for numerical columns
for col in numerical_cols:
    plt.figure(figsize=(8, 5))
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')
    plt.xlabel(col)
    plt.show()
```

```python
#7: Data Visualization - Bar Chart Analysis for Categorical Columns
for col in categorical_cols:
    plt.figure(figsize=(8, 5))
    sns.countplot(x=df[col])
    plt.title(f'Bar Chart of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.show()
```

```python
# 8: Standardization of Numerical Variables - Z-score normalization
standardized_df = df[numerical_cols].apply(zscore)
standardized_df.columns = [f'{col}_z' for col in numerical_cols]

# Show before and after comparison of data distributions
for col in numerical_cols:
    fig, axes = plt.subplots(1, 2, figsize=(15, 5))
    sns.histplot(df[col], kde=True, bins=30, ax=axes[0])
    axes[0].set_title(f'Original Distribution of {col}')
    sns.histplot(standardized_df[f'{col}_z'], kde=True, bins=30, ax=axes[1])
    axes[1].set_title(f'Standardized Distribution of {col}')
    plt.show()
```

```python
# 9: Conversion of Categorical Data into Dummy Variables
# One-hot encoding
encoded_df = pd.get_dummies(df[categorical_cols], drop_first=True)

# Display a portion of the transformed dataset
encoded_df.head()
```

```python
#  10: Combine standardized numerical and encoded categorical data
processed_df = pd.concat([standardized_df, encoded_df], axis=1)

# Display the first few rows of the processed dataset
processed_df.head()
```

```python
# Save the processed dataset to a new CSV file
processed_df.to_csv('processed_sales_discounts.csv', index=False)
```
